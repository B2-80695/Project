## DATA PREPROCESSING

### Deleting Unnecessary Attributes

```python
df.drop(['date_time', 'moonrise', 'moonset', 'sunrise', 'sunset'], axis = 1, inplace = True)
```

### Finding The Missing Data (NaN Values)

```python
df.isna().sum()
```

### Filling The Missing Data

```python
df.mean()
```

```python
filled_df = df.fillna(df.mean())
```

```python
filled_df.isna().sum()
```

### Scaling The Data

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(filled_df)
new_df = scaler.transform(filled_df)
```

```python
DF = pd.DataFrame(new_df, columns = df.columns)
```
